```python
import matplotlib.pyplot as mpl
import matplotlib.pylab as plt
import numpy as np

mpl.rc("font" , family="NanumGothic")
mpl.rc("axes", unicode_minus = False)
```

```python
colors = ["blue", "green", "red"]
y = [2, 3, 1]
x = np.arange(len(y)) ## [0,1,2]

xlabel = ["가","나", "다"]
plt.title("Bar Chart")
plt.bar(x,y)
plt.yticks(sorted(y))
plt.yticks(y)
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(0)

people = ["몽룡", "춘향", "방자", "향단"]
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))
print(performance)
print(error)

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel("x 라벨")
plt.show()

```

```python
x = np.linspace(0.1, 2 * np.pi,10)

plt.title("Stem Plot")
plt.stem(x, np.cos(x), "-.")
plt.show()
```

```python
#파이 차트
labels = ["개구리", "돼지", "개", "통나무"]
sizes = [20, 30, 45, 10]
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral"]
explode = (0.1, 0.1, 0.2, 0) # 원에서 얼만큼 떨어지느냐
plt.title("Pie Chart", fontdict = {"fontsize" : 20})
plt.pie(sizes, explode=explode, labels = labels, colors = colors,
       autopct = "%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.show()
```

```python
#히스토 그램

np.random.seed(0) 
x = np.random.randn(1000) #표준정규분포
plt.title("Histogram")
arrays,bins,paches = plt.hist(x, bins = 100)
plt.show()
```
